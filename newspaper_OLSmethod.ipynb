```python
import statsmodels.formula.api as sm
```

```python
formula = "sunday ~ daily"
```

```python
model=sm.ols(formula = formula, data=data)
```

```python
model1=model.fit()
```

```python
model1.params
```

```python
model1.summary()
```

```python
newdata=pd.Series([200,300])
```

```python
data_pred=pd.DataFrame(newdata,columns=['daily'])
```

```python
data_pred
```

```python
model1.predict(data_pred)
```
